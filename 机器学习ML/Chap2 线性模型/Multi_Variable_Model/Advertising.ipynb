```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from Multi_Linear_Regression import Linear_Regression_Multi_Variable
```

```python
# 加载模型
Advertisment = Linear_Regression_Multi_Variable()
```

```python
# 加载数据
filepath = "../../dataset/Advertising.csv"
Advertisment.Dataloader(filepath)
```

```python
# 划分数据集合
Advertisment.Train_test_split(seed=1)
```

```python
# 计算系数和各个参数的指标
Advertisment.regression_coefs()
Advertisment.statistical_magnitude()
```

```python
coefs = Advertisment.coef_
print("Sales = {:.4f} + {:.4f} * TV + {:.4f} * Radio + {:.5f} * Newspaper".\
      format(coefs[-1], coefs[0], coefs[1], coefs[2]))
print("RMSE = {:.4f}".format(Advertisment.rmse))
print("r2 = {:.4f}".format(Advertisment.r2_score))
```

```python
# 计算测试集合上的y值
pred_y = Advertisment.predict(Advertisment.test_x)
```

```python
# 画图
%config InlineBackend.figure_format = "svg"
#plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif']=['SimSun']
plt.rcParams['axes.unicode_minus'] = False
x_num = np.linspace(1, len(pred_y), len(pred_y))
plt.plot(x_num, pred_y, 'b*--', label="回归测试值")
plt.plot(x_num, Advertisment.test_y, 'r+-', label="真实值")
plt.title("多元线性回归基于测试值的预测效果图")
plt.xlabel("测试样本序号")
plt.ylabel("测试样本Sales观测值")
plt.grid(linestyle="-.")
plt.legend()
plt.savefig("Multi_test.svg")
plt.show()
```
